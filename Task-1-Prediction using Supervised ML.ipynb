<h2>Problem Statement: Predict the perecentage of the student based on the no. of hours student studied using Simple Linear Regression model which involves two variables</h2>

```python
# Importing the required libraries
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h2>Reading Data From the source</h2>

```python
# Reading data from remote link
stu_data = pd.read_csv(r"http://bit.ly/w-data")
print("Data imported successfully")

stu_data.head()
```

<h2>Visualizing Data </h2>
<h3>By visualizing data we can find correlation between Hours and Percentage </h3>

```python
# Plotting the distribution of scores
stu_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage ')  
plt.show()
```

```python
X = stu_data.iloc[:, :-1].values  
y = stu_data.iloc[:, 1].values
```

<h2>Splitting the data into training and testing sets</h2>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

<h2>Model Training</h2>

```python
regressor = LinearRegression()  
regressor.fit(X_train.reshape(-1,1), y_train) 

print("Model Trained.")
```

<h2> Visualize the best-fit line of regression</h2>

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line,color='red');
plt.show()
```

<h2>Testing Data and Model Prediction</h2>

```python
# Testing data
print(X_test)
# Model Prediction 
y_pred = regressor.predict(X_test)
```

<h2>Comparing Actual vs Predicted Value </h2>

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```

```python
#Estimating training and test score
print("Training Score:",regressor.score(X_train,y_train))
print("Test Score:",regressor.score(X_test,y_test))
```

<h2>Comparing Actual vs Predicted Value Graphically </h2>

```python
# Plotting the Bar graph to depict the difference between the actual and predicted value

df.plot(kind='bar',figsize=(5,5))
plt.grid(which='major', linewidth='0.5', color='red')
plt.grid(which='minor', linewidth='0.5', color='blue')
plt.show()
```

```python
# Testing the model with given amount of data that is 9.25
hours = 9.25
test = np.array([hours])
test = test.reshape(-1, 1)
own_pred = regressor.predict(test)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

<h2>Model Evaluation</h2>

```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R^2:', metrics.r2_score(y_test, y_pred))
```

<p><bl>In regression, the R2 coefficient of determination is a statistical measure of how well the regression predictions approximate the real data points.<br></bl>
R-squared is a statistical measure of how close the data are to the fitted regression line.Here R^2 value is 0.945 which is close to 1 so it is a good score for such a less amount of data.</p>
